```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stocks = pd.read_csv('stock_px.csv', parse_dates=True, index_col=0)
stocks
```

```python
plt.title('stock')
plt.plot(stocks['MSFT'], label='MSFT')
plt.plot(stocks['XOM'], label='XOM')
plt.legend(loc='best')
plt.show()
```

```python
fig, axs = plt.subplots(4,1) 
fig.set_figheight(10)
fig.set_figwidth(15)

axs[0].set_title('AAPL')
axs[0].plot(stocks['AAPL'])

axs[1].set_title('MSFT')
axs[1].plot(stocks['MSFT'])

axs[2].set_title('XOM')
axs[2].plot(stocks['XOM'])

axs[3].set_title('SPX')
axs[3].plot(stocks['SPX'])

plt.tight_layout() # 여러 그래프 정열 (margin을 가장 작게 채워서 사용함)
plt.show()
```
