```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.DataFrame({

    '학교': ['서울대', '서강대', '부산대', '충남대', '경북대'],

    '지역': ['서울', '서울', '부산', '대전','대구'],

    '학생수':['10,712','8,179','11,935','10,991','11,314'],

    '학과 수': ['56', '45', '58','51','59'],

    '평균등록금' : ['2,110,000','5,340,000','2,430,000','2,347,000','2,522,000']}, columns=['학교', '지역', '학생수', '학과 수', '평균등록금'])

df_2 = pd.DataFrame({

    '학교평가등급': ['B+', 'A', 'A', 'A-', 'A+'],

    '학생만족도':['81점','79점','77점','84점','82점'],

    '취업자수': ['5,122명', '4,987명', '5,001명','5,117명','4,777명'],

     'area': ['서울', '서울', '부산', '대전','대구'],

    '면적' : ['9,122','7,101','9,095','8,873','9,002']

})

df_1
```

```python
df_2
```

```python
#  a. 위의 두 데이터프레임의 공통 키(지역, 'area')를 이용하여 merge 명령어로 합친다
a= pd.merge(df_1, df_2, left_on = '지역', right_on='area')
```

```python
# b. 위의 데이터프레임들에서 중복되는 키 이름(지역, 'area') 중 하나는 제거하고 출력한다.
b= pd.merge(df_1, df_2, left_on = '지역', right_on='area').drop(columns='area')
b
```

```python
df1 = pd.DataFrame({

    '매출' : [10000000, 12000000, 9000000, 6000000, 8000000, 1100000],

    '비용' : [15000000, 1300000, 1200000, 9000000, 9900000, 9500000]},

    index = ['1월', '2월', '3월', '4월', '5월', '6월'])



df2 = pd.DataFrame({

    '매출' : [13000000, 14000000, 17000000, 15400000, 16500000, 16600000],

    '비용' : [11000000, 10400000, 11000000, 12100000, 9000000, 9500000]},

    index = ['7월', '8월', '9월', '10월', '11월', '12월'])
    
df1
```

```python
df2
```

```python
# 실적 정보는 "매출", "비용", "이익" 으로 이루어진다. (이익 = 매출 - 비용). 
# 또한 1년간의 총 실적을 마지막 행으로 덧붙인다.
df3 = pd.concat([df1, df2], axis=0)
df3['이익'] = df3['매출']-df3['비용']
df3.loc['실적'] = pd.Series([df3['매출'].sum(), df3['매출'].sum(), df3['매출'].sum()], 
                               index=['매출', '비용', '이익'])
df3
```
