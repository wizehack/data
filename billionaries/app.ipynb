```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data/dollar_billionaires_per_million_people.csv')
data
```

```python
data['billionaires_per_million_inhabitants'] = data['billionaires_per_million_inhabitants'].str.upper()
data['avg'] = data.loc[:,'2004':'2007'].mean(axis=1)
data['rank'] = data['avg'].rank(ascending=False)
data.sort_values(by='avg', ascending=False).head(10)
data
```

```python
condition = data.billionaires_per_million_inhabitants.str.contains('KOREA, REP.')
data[condition]
```

```python
data[condition].loc[:,'2004':'2007'].plot(kind='bar', xlabel='billionaires_per_million_inhabitants')
```
