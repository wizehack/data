```python
'''
1. 와인 품질 데이터셋(winequality-both.csv)을 이용한다.

 a. 데이터 셋을 DataFrame으로 읽어온다.

 b. 변수별 요약 통계를 표시한다.(최소, 최대, 사분위, 평균, 표준편차)

 c. 와인 품질(quality)의 유일 값(점수)을 찾아 출력한다.

 d. 와인 품질(quality)의 빈도(점수 별 갯 수)를 계산하여 출력한다.

 e. 와인 종류(type)에 따른 (alcohol)의 기술 통계(describe())를 출력한다. 

 f.  와인 종류에 따른 품질(quality)의 분포를 확인하기 위해 레드와인과 화이트 와인의 품질 값(quality)을 막대 그래프로 출력한다.
    (이때 범례도 같이 출력한다. )

 g. 와인 종류에 따라 와인 품질 값(quality)의 표준편차(std)와 평균(mean)값을 출력한다.
'''
```

```python
# a. 데이터 셋을 DataFrame으로 읽어온다.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

frame = pd.read_csv('winequality-both.csv')

frame
```

```python
# b. 변수별 요약 통계를 표시한다.(최소, 최대, 사분위, 평균, 표준편차)
frame.describe()
```

```python
#c. 와인 품질(quality)의 유일 값(점수)을 찾아 출력한다.
frame['quality']
```

```python
# d. 와인 품질(quality)의 빈도(점수 별 갯 수)를 계산하여 출력한다.
frame['quality'].value_counts()
```

```python
#e. 와인 종류(type)에 따른 (alcohol)의 기술 통계(describe())를 출력한다. 
#frame.pivot_table('alcohol', columns=['type']).describe()

frame.pivot_table(['alcohol'], index=['type'], aggfunc='describe')
```

```python
'''
f.  와인 종류에 따른 품질(quality)의 분포를 확인하기 위해 레드와인과 화이트 와인의 품질 값(quality)을 막대 그래프로 출력한다.
    (이때 범례도 같이 출력한다. )
'''

df = frame[['quality','type']].value_counts().unstack()
df.plot(kind='bar')
plt.show()
```

```python
'''
g. 와인 종류에 따라 와인 품질 값(quality)의 표준편차(std)와 평균(mean)값을 출력한다.
'''

gf = frame[['quality','type']].value_counts().unstack()
gf.agg(['mean', 'std'])
```
