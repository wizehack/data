```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fec = pd.read_csv('P00000001-ALL.csv')
fec.info()
```

```python
data1 = fec[['cand_nm', 'contbr_nm','contbr_employer', 'contb_receipt_amt', 'contb_receipt_dt']]
```

```python
data1.pivot_table('contb_receipt_amt',
                                index='contbr_employer', 
                                aggfunc='sum').sort_values(by='contb_receipt_amt', ascending=False )
```

```python
data1.pivot_table('contb_receipt_amt', 
                                index='cand_nm', 
                                aggfunc='sum').sort_values(by='contb_receipt_amt', ascending=False )
```
